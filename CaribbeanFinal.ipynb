### I. Region introduction

The Caribbean region comprises 19 island nations with beautiful beaches, vibrant cultures, and unmatched biodiversity. Its strategic location and abundant natural resources have caused European colonialism to play a significant role in the region since the fifteen century. This accounts for Spanish, French, English, and Dutch languages spoken throughout each respective country. Six of the nineteen countries are still considered territories of foreign nations, including Aruba, Cayman Islands, Puerto Rico, St Kitts and Nevis, British Virgin Islands, and the Virgin Islands. 

These islands have lucrative economies, with GNI per capita ranging from 800 to 30k, and mainly rely on tourism, commodity exports, agriculture, and oil production. Recently, these countries have been increasingly invested in cultivating a “blue economy,” that enables sustainable use of marine resources and environmentally sound infrastructure in the energy sector (World Bank, 2020).

The region has been increasingly plagued by extreme environmental disasters like hurricanes and earthquakes, which have decimated a few Caribbean islands over the past 5 years, especially in Puerto Rico where over 3,000 people died due to Hurricane Maria in 2017 alone (Florido, 2019). This area is extremely vulnerable to the consequences of global warming. 

The Caribbean nations in general have made substantial economic and social growth since 2010, however, chronic poverty is still rampant in the region largely due to ramifications of colonialism, neoliberalism, political instability, and environmental disasters (Vakis, Rigolini, & Luchetti, 2015)

### Outliers

The team found that there were many outliers within the Caribbean data set, and created boxplots for all columns to determine how outliers were distributed. When we completed that task, the team categorized the 41 variables into 5 overall categories (education, standard of living, economy, health, and industrial production). Then, we selected the most significant variable within each category to give a holistic representation of Caribbean through analysis on the Dominican Republic.

After analyzing the figures for the top 5 categories, we checked their threshold values to remove the outliers from our dataset. We flagged the outliers from our data and then created new columns for them to be stored, which enabled us gain a deeper and more accurate understanding of the data. 

```python
# Code for outliers


for x in score_four:
    fig, ax = plt.subplots(figsize = (20, 6))
    sns.boxplot( x = x, 
              y = None, 
              hue    = None,     
              orient = 'h',     
              data   = ratchet)
    
    ax.set_xlabel(x, 
              fontdict={'fontsize':12})
    
    plt.show()
    
    
# Top 5 Features’ thersholds

#'lit_rate', <87
#'maternal_mort', >95
#'emissions', >11.05
#'gdp_emp',
#'water_source', <92.5
#‘transformer’ == ‘Ratchet’

########################
# For the variable lit_rate
ratchet['lit_rate_outlier'] = 0

# Setting threshold
lit_threshold = 87

# looping to change water_outlier
for index, lit in ratchet.loc[ : , : ].iterrows():
    if ratchet.loc[ index , 'lit_rate'] < lit_threshold:
        ratchet.loc[ index , 'lit_rate_outlier'] = 1

# Printing result
print(ratchet.loc[ : , 'lit_rate_outlier'].value_counts())



########################
# For the variable maternal_mort
ratchet['maternal_mort_outlier'] = 0

# Setting threshold
maternal_threshold = 95

# looping to change maternal_mort_outlier
for index, mat in ratchet.loc[ : , : ].iterrows():
    if ratchet.loc[ index , 'maternal_mort'] > maternal_threshold:
        ratchet.loc[ index , 'maternal_mort_outlier'] = 1

# Printing result
print(ratchet.loc[ : , 'maternal_mort_outlier'].value_counts())



########################
# For the variable emissions
ratchet['emissions_outlier'] = 0

# Setting threshold
emissions_threshold = 11.05

# looping to change emissions_outlier
for index, emissions in ratchet.loc[ : , : ].iterrows():
    if ratchet.loc[ index , 'emissions'] > emissions_threshold:
        ratchet.loc[ index , 'emissions_outlier'] = 1

# Printing result
print(ratchet.loc[ : , 'emissions_outlier'].value_counts())


########################
# For the variable gdp_emp
# No outlier for this variable


########################
# For the variable water
ratchet['water_outlier'] = 0

# Setting threshold
water_threshold = 92.5

# looping to change water_outlier
for index, water in ratchet.loc[ : , : ].iterrows():
    if ratchet.loc[ index , 'water_source'] < water_threshold:
        ratchet.loc[ index , 'water_outlier'] = 1

# Printing result
print(ratchet.loc[ : , 'water_outlier'].value_counts())
```

   Literacy Rate adult total (percent of people ages 15 and above) provides insight into the overall effectiveness of the education systems and quality of labor skills for both men and women. Throughout the world, literacy rates range from 46.53% to 97.64%, and the Caribbean ranks third highest with an average score of 95.94%.    Maternal mortality ratio (modeled estimate, per 100,000 live births) is an indicator of access to adequate health care facilities and socioeconomic status of a country. The Caribbean region has the 5th lowest mortality rates in the world with a 91 out of 100,000 birth mortality rates, compared to 8 in North America and 583 in Sub-Saharan Africa. In the Dominican Republic, 85 women, on average, due during childbirth per 100,000 births.

  Team 1 used CO2 emissions (metrics ton per capita) to analyze the Caribbean’s industrialization and contribution to global warming.  As mentioned in the introduction, the Caribbean islands have put concerted efforts towards building a Blue Economy. However, one of the most surprising findings is that the Caribbean ranks the third highest in CO2 emissions out of the 12 regions of the world with the average of 7.09 ton per capita. Although the Dominican Republic has a value of 2. 15, it is still a strong representation of CO2 emissions among the small Caribbean islands due to two major outliers, Aruba and Trinidad and Tobago, which skew the distributions to the right. Oil production and cement manufacturing for infrastructure development cause drastic differences in averages (Andersen, 2020). The overall CO2 emission for the Caribbean states are relatively small in comparison with a range from 0.24 to 6.30, omitting the outliers. 

   GDP per person employed (constant 2011 PPP in USD) was chosen as a key feature to analyze because some data points correlate to higher rates of poverty (for example, moderately high adolescent fertility rates and prevalence of undernourishment percentages). However, the team’s analysis show that the Caribbean has the fifth highest GDP per PPP out of the 12 regions of the world with USD 42,605.30 per person employed. It is interesting to discover in the data that these small Caribbean islands have some of the highest GDP per capita in the world. Updated data from the World Bank exemplifies the Caribbean’s overall lucrative economy, which is represented in the Dominican Republic with USD 42,070 GDP per person (World Bank, 2020).

   Access to improved water source is a significant sign of a country’s sanitization and public health standards. The Caribbean ranks third clean water source with 93.33 percent of the population allowed access. Countries in the region have invested in sustainable development as illustrated by the Dominican Republic’s 86.31 percent. It is interesting to note the outliers in this case which are and Puerto Rico both with less than 59 percent of the population having access to water.
